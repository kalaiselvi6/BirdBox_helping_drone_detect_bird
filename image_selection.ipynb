# Manually select image for evaluating performance
Only those images similar to images shot from drone (i.e., flapping wing moment) will be used for evaluation.

```python
from IPython import display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from glob import glob
import os
```

```python
# getting all the images 
IMAGE_PATH = r'E:\Download\TensorflowObjectDetection\TFODCourse\Tensorflow\workspace\images\test'
images = list(set(glob(os.path.join(IMAGE_PATH, '*'))) - set(glob(os.path.join(IMAGE_PATH, '*.xml'))))
```

```python
def select_image(i):
    '''
    Copy image from test folder to select_birds folder. The function will display image and ask for user input; 'y' or '3' yes, 'e' exit the function, and any other key or just simply press enter to move to the next image in the list.

    INPUT:
    i : [int] the index for image from images array
    '''
    a = None
    
    plt.figure()
    image_object = mpimg.imread(images[i])
    plt.imshow(image_object)

    # user plt.show() so that at the end notebook will not displayy all the photos
    plt.show()
    # display.display(plt.gcf())

    # keeping track of where we are in the list
    print('Currently at: ', i)
    print(images[i])

    # prompt for user pinput
    a = input('select ? (y/n/e)')
    if a.lower() in ['y', '3']: # 3 is close to my right enter, so i can operate 1 hand 
        os.system(f"copy {images[i].split('.')[0] +r'.*'} I:\\select_birds")
    if a.lower() == 'e':
        return None
    
    # clear output, so that next image can be displayed in place of the current one
    display.clear_output(wait=True)

    # move onto the next image in the list images
    return select_image(i+1)
```
